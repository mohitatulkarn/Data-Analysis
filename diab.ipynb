# outlier detection

```python
## using iqr
```

```python
Q1 = df1.quantile(0.25)
```

```python
Q3 = df1.quantile(0.75)
```

```python
IQR = Q3-Q1
```

```python
df2 = df1[~((df1<(Q1-IQR *1.5))|(df1>(Q3+IQR*1.5))).any (axis=1)]
```

```python
df2.shape
```

```python
df.shape
```

```python
i=list(df1.loc[:,df1.dtypes !=object].columns)
```

```python
from scipy import stats
```

```python
z =np.abs(stats.zscore(df1[i]))
```

```python
print(z)
```

```python
thresold = 3
```

```python
df3 = df1[(z<thresold).all(axis=1)]
```
