```python
#age -survival
sns.boxplot(y='age',x='survival',data=df)
```

#here age>77 will not live more than 5 years
#herea age<33 will live more than 5 years

```python
sns.displot(x='age',kde=True,bins=4,data=df,hue='survival')
```

```python
#age and nodes
fig,ax= plt.subplots(figsize=(10,6))
ax.scatter(x='age',y='Nodes',data=df)
plt.show()
```

here we can see that the number of nodes between age 33 to 70 is usually greater than 5
age<33 have less than  4 nodes

```python
fig = plt.subplots(figsize =(10,8))
sns.barplot(x='age',y='Nodes',data=df)
plt.show()
```

```python
#year-survival
sns.countplot(x='year',hue='survival',data=df)
```

here we can see in year 1958 an 1960 have highest number of people who live more than 5 years
and in year 1965 have highest no of people who live less than 5 years



```python
#node and survival
sns.boxplot(y='Nodes',x='survival', data=df)
plt.show()
```
