```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df= pd.read_csv("C:\\Users\\AtulKumar\\Downloads\\social media data.csv")
```

```python
df.head()
```

```python
def fucc (x):
    if(0<x<11):
        return "0-10"
    if (10<x<16):
        return "11-15"
    if (15<x<21):
        return "16-20"
    if (20<x<26):
        return "21-25"
    if (25<x<31):
        return "26-30"
    if(30<x<41):
        return "30-40"
    if (40<x<51):
        return "41-50"
    if(50<x<61):
        return "51-60"
    else:
        return "60<"
df['age_group']= df['age'].apply(fucc)
```
